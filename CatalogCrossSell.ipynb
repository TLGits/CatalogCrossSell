```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_excel('CatalogCrossSell.xls')
# data = data.set_index('Customer Number')
data.drop(columns = 'Customer Number', inplace = True)
```

```python
data
```

```python
apr_result = apriori(data, min_support = 0.1, use_colnames = True)
```

```python
fpg_result = fpgrowth(data, min_support = 0.1, use_colnames = True)
```

```python
apr_result
```

```python
fpg_result
```

```python
apr_con_asso = association_rules(apr_result, metric="confidence", min_threshold=0.7)
apr_con_asso
```

```python
fpg_con_asso = association_rules(fpg_result, metric="confidence", min_threshold=0.7)
fpg_con_asso
```

```python
apr_lift_asso = association_rules(apr_result, metric="lift", min_threshold=1.4)
apr_lift_asso
```

```python
fpg_lift_asso = association_rules(fpg_result, metric="lift", min_threshold=1.4)
fpg_lift_asso
```
